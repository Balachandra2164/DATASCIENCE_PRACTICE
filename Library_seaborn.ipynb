```python
from numpy.random import randn
```

```python
np.random.seed(100)
```

```python
df=pd.DataFrame(dict(time=np.arange(500),value=randn(500).cumsum()))
```

```python
df.head()
```

```python
sns.relplot(x='time',y='value',data=df,kind='line',sort=True)
```

```python
df1=pd.DataFrame(randn(500,2).cumsum(axis=0),columns=['time','value'])
```

```python
df1.head()
```

```python
sns.relplot(x='time',y='value',data=df1,kind='line',sort=True)
```

```python
fmri=sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
fmri.shape
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri)
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,kind='line')
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',ci='sd')
#when you use standard deviation it actualshows actual region 
```

```python
sns.relplot(x='timepoint',y='signal',estimator=None,data=fmri,kind='line',ci='sd')
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,hue='event',kind='line')
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,hue='event',kind='scatter')
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,hue='region',style='event',kind='line')
#fmri.info()
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,hue='region',style='event',kind='line',markers=True,dashes=True)
#fmri.info()
```
